```python
#Import Path and csv library
from pathlib import Path
import csv
```

```python
#Establish file path
file_path = Path('Resources/budget_data.csv')

#Initialize line_num variable
line_num = 0

#Initialize company_pnl and date to hold values in csv file
company_pnl = []
dates = []
```

```python
#Open input path as file object
with open(file_path, 'r') as csvfile:
    
#Print data type    
    print(type(csvfile))
    
#Push file through csvreader and establish delimiter    
    csvreader = csv.reader(csvfile, delimiter=',')
    
#Print data type of csvreader  
    print(type(csvreader))
    
#Establish next line as header
    header = next(csvreader)
    line_num += 1
    
#Print header
    print(f'{header}')
    
#Read row of data after the header
    for row in csvreader:
        print(row)
        
#Establish the PnL column as budget
        budget = (int(row[1]))
        date = (str(row [0]))
#Add the values in budget to list called company_pnl
        company_pnl.append(budget)
        dates.append(date)
```

```python
#Initialize relevant variables
total_months = 0
total_pnl = 0
avg_change = 0
high_change = 0
low_change = 0
```

```python
#TOTAL MONTHS & TOTAL PNL
for pnl in company_pnl:
    total_months += 1
    total_pnl += pnl
```

```python
#AVERAGE CHANGE
#Initialize appropriate lists to use zip() method to join values
total_change = 0
change_count = 0
monthly_change = []

#Create variables where company_pnl is staggered. This allows the value at T to be subtracted by (T-1)
pnl_1 = company_pnl[0:]
pnl_2 = company_pnl[1:]

#Connect the values in the lists to each other (pnl_1[0] to pnl_2[0], pnl_1[1] to pnl_2[1])
pair_lists = zip(pnl_1, pnl_2)

#Loop both paired values in the lists and add the value to monthly_change list
for pnl_1_i, pnl_2_i in pair_lists:
    monthly_change.append(pnl_2_i-pnl_1_i)

#Loop new monthly_change list and sum all values and determine how many values
for change in monthly_change:
    total_change += change
    change_count += 1

#Establish avg_change equation with new values for change
avg_change = round(total_change / change_count, 2)
```

```python
#MAX & MIN CHANGE
#Loop monthly_change list
for pnl in monthly_change:
#If statement to establish low_change value    
    if low_change == 0:
        low_change = pnl

#Find lowest value in monthly_change
    elif pnl < low_change:
        low_change = pnl

#Find highest value in monthly_change
    elif pnl > high_change:
        high_change = pnl
```

```python
#Find date for high_change and low_change
#Add 1 to index to show when the change was actually seen
high_date = (monthly_change.index(high_change)) + 1
low_date = (monthly_change.index(low_change)) + 1
```

```python
#Print Financial Analysis
print('Financial Analysis')
print('---------------------------------------')
print(f'Total Months: {total_months}')
print(f'Total: ${total_pnl}')
#For formatting reasons, this if else statement puts the negative symbol in front of the $
if avg_change < 0:
    print(f'Average Change: -${abs(avg_change)}')
elif avg_change > 0:
    print(f'Average Change: ${avg_change}')
print(f'Greatest Increase in Profits: ${high_change}')
print(f'Greatest Decrease in Profits: -${abs(low_change)}')
```
